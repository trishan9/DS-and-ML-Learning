```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Position_Salaries.csv")
dataset
```

```python
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
print(X, y)
```

```python
from sklearn.linear_model import LinearRegression

linear_regressor = LinearRegression()
linear_regressor.fit(X, y)
```

```python
plt.scatter(X, y, color='green')
plt.plot(X, linear_regressor.predict(X), color='red')
plt.title("Linear Regression Model for Level v/s Salary")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()

print(f"Predicted Salary for Level 7.5: {linear_regressor.predict([[7.5]])}")
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

pf = PolynomialFeatures(degree=5)
X_poly = pf.fit_transform(X)

polynomial_regressor = LinearRegression()
polynomial_regressor.fit(X_poly, y)
```

```python
plt.scatter(X, y, color='blue')
plt.plot(X, polynomial_regressor.predict(X_poly), color='red')
plt.title("Polynomial Regression Model for Level v/s Salary")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()

print(f"Predicted Salary for Level 7.5: {polynomial_regressor.predict(pf.fit_transform([[7.5]]))}")
```
